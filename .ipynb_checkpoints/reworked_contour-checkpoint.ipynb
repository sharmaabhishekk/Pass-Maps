```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('final_bissouma.csv')
df.head()
```

```python
from mplsoccer.pitch import Pitch
import seaborn as sns

pitch = Pitch(orientation='vertical', view='full', figsize=(8, 16), pitch_type='opta')

fig, ax = pitch.draw()
sns.kdeplot(df.y, df.x, ax=ax, cmap='plasma', shade=True, alpha=0.5, zorder=1)
plot_comet = lambda vals: ax.annotate('', xy=(vals[0], vals[1]), xycoords='data', xytext=(vals[2], vals[3]), 
                                     arrowprops=dict(arrowstyle="wedge,tail_width=0.3", fc="0.2",
                                                     ec="none"), zorder=2
                                    )

scat = ax.scatter(df['endY'], df['endX'], c=df['xT'], cmap='viridis', zorder=10)

df[['y', 'x', 'endY', 'endX']].apply(plot_comet, axis=1);
plt.colorbar(scat, shrink=.2)
```
